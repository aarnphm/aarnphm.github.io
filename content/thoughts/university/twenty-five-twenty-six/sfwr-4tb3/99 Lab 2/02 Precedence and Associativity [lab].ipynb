#### Precedence and Associativity

Programmers and compilers can make programs more efficient by eliminating duplicate computations.

- Operator `+` associates to the left means `a + b + c` is parsed as `(a + b) + c`.
- Operator `+` associates to the right means `a + b + c` is parsed as `a + (b + c)`.
- Operator `+` is associative means `(a + b) + c = a + (b + c)`, i.e. it does not matter how it is parsed.
- Operator `+` is commutative means `a + b = b + a`.
