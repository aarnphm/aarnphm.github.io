### Call by Name

Consider the following program:
```Pascal
var s: integer
procedure sum(i, l, h, x: integer)
  s := 0; i := l
  while i < h do
    s := s + x; i := i + 1
```

If all parameters are passed by name, what is the final value of `s` after the call
```Pascal
sum(a, 1, n, a × a)
```
assuming that `a` and `n` are integer variables? Use the definitions to argue!

By the definition of call by name, each formal parameter in the procedure body is textually replaced by the actual parameter expression, evaluated in the caller's scope each time it is referenced.

Call: `sum(a, 1, n, a × a)` substitutes i → a, l → 1, h → n, x → a × a:

```
s := 0
a := 1              (i := l)
while a < n do      (i < h)
  s := s + a × a    (s := s + x, where x = a × a is re-evaluated each iteration)
  a := a + 1        (i := i + 1)
```

The parameter `x` is bound to the expression `a × a`, not its initial value. Since `a` is modified by `i := i + 1` (which becomes `a := a + 1`), the value of `x` changes each iteration:

- a = 1: s = 0 + 1×1 = 1, then a = 2
- a = 2: s = 1 + 2×2 = 5, then a = 3
- a = 3: s = 5 + 3×3 = 14, then a = 4
- ...
- a = n−1: s = ... + (n−1)², then a = n
- Loop exits: a < n is false.

**Final value: s = 1² + 2² + 3² + ... + (n−1)² = (n−1)·n·(2n−1) / 6.**

This is possible only with call by name: the expression `a × a` acts as a thunk that captures the current value of `a` each time `x` is referenced, producing different values across iterations.
