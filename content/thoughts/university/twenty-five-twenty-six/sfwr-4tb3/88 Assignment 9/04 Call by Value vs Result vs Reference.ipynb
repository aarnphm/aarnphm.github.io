### Call by Value vs Result vs Reference

Consider the following program:
```Pascal
var a, b: integer
procedure q(x: integer)
  x := a + 1; b := a
```

Suppose initially `a = 1` and `b = 3`. What are the final values of `a` and `b` after the call
```Pascal
q(a)
```
if the sole parameter of `q` is passed by value, by result or by reference? Use the formal definitions to argue!

Given `a = 1`, `b = 3`. Call `q(a)` where `procedure q(x: integer): x := a + 1; b := a`.

**Call by value:** By definition, the actual parameter is evaluated and the value is assigned to a new local variable x. So x = 1 (copy of a). Execute: `x := a + 1 = 1 + 1 = 2` (this modifies x only; a remains 1). `b := a = 1`. **Final: a = 1, b = 1.**

**Call by result:** By definition, x is a new local variable (uninitialized). On procedure exit, x is copied to the actual parameter a. Execute: `x := a + 1 = 1 + 1 = 2` (a is still 1 during execution since x is local). `b := a = 1`. On return: `a := x = 2`. **Final: a = 2, b = 1.**

**Call by reference:** By definition, x becomes an alias for a (bound to a's address). Any modification to x directly modifies a. Execute: `x := a + 1 = 1 + 1 = 2`. Since x IS a, this sets a = 2 immediately. `b := a = 2` (a was already changed). **Final: a = 2, b = 2.**

| | a | b |
|---|---|---|
| by value | 1 | 1 |
| by result | 2 | 1 |
| by reference | 2 | 2 |

The difference between result and reference is visible in the second statement `b := a`: under call by reference, the assignment `x := a + 1` has already changed a (since x aliases a), so b sees the updated value 2. Under call by result, x is a separate local, so a is unchanged until the procedure returns.
