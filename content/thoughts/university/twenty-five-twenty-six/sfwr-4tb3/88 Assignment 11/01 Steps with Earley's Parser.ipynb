#### Steps with Earley's Parser

Consider Earley's parser from the course notes modified to print a markdown table:

```python
def parse(g: tuple[str, ...], x: str, log = False):
    global s
    n = len(x); x = '^' + x + '$'; S, ŌÄ = g[0][0], g[0][2:]
    k, s = 0, [{(S, '', ŌÄ, 0)}] + [set() for _ in range(n)]
    if log: print('| ', k, ' |   | s[0] | ', S, '‚Üí ‚ÄĘ', ŌÄ, ', 0 |', sep='')
    for i in range(n + 1):
        v = set() # visited items
        while v != s[i]:
            e = (s[i] - v).pop(); v.add(e) # pick an arbitrary un-visited item
            A, ŌÉ, ŌĄ, j = e
            if len(ŌĄ) > 0 and ŌĄ[0] == x[i + 1]: # match, a == ŌĄ[0]
                f = (A, ŌÉ + ŌĄ[0], ŌĄ[1:], j)
                s[i + 1].add(f); k += 1
                if log: print('| ', k, ' | M | s[', i + 1, '] | ', f[0], '‚Üí', f[1], '‚ÄĘ', f[2], ', ', f[3], ' |', sep='')
            elif len(ŌĄ) > 0: # predict, B == ŌČ[0]
                for f in ((r[0], '', r[2:], i) for r in g if r[0] == ŌĄ[0]):
                    s[i].add(f); k += 1
                    if log: print('| ', k, ' | P | s[', i, '] | ', f[0], '‚Üí', f[1], '‚ÄĘ', f[2], ', ', f[3], ' |', sep='')
            else: # complete, len(ŌĄ) == 0
                for f in ((B, őľ + őĹ[0], őĹ[1:], k) for (B, őľ, őĹ, k) in s[j] if len(őĹ) > 0 and őĹ[0] == A):
                    s[i].add(f); k += 1
                    if log: print('| ', k, ' | C | s[', i, '] | ', f[0], '‚Üí', f[1], '‚ÄĘ', f[2], ', ', f[3], ' |', sep='')
    return (S, ŌÄ, '', 0) in s[n]
```
