When cells contain commas (`,`), we need to wrap them in double quotation marks (e.g., `"..."`), double-quoting all contained double-quotation marks too (e.g., `A cell, like this one, containing "double quotes".` becomes `"A cell, like this one, containing ""double quotes""."`). For `data/excessive-quotes.csv`, write a conservative `sed` command that removes double quotation marks surrounding cells that contain neither commas nor double-quotation marks. With `sed`, you may use a loop (format: `loop_label: s/.../.../g; t loop_label`) to ensure that substitutions are applied as many times as needed on a line until all possible substitutions are applied.

```python
%%capture output
%%bash
sed -E ':loop; s/(^|,)"([^",]*)"(,|$)/\1\2\3/g; t loop' excessive-quotes.csv
```
