Generate the RISC-V assembly code of the following P0 program:

```python
compileString("""
var s: integer
var a: [0 .. 9] → integer
procedure sum(l, h: integer)
  var i: integer
    s := 0; i := l
    while i ≤ h do
        s := s + a[i]; i := i + 1
program printsum
  a[0] := 5; a[1] := 7
  sum(0, 1)
  write(s)
""", 'printsum.s', target = 'riscv')
```

You can view the RISCV assembly code:

```python
!cat printsum.s
```

You can execute RISC-V code with an RISC-V emulator, like the Python `riscemu` library. As the emulator reads from `stdin`, P0 programs that call `read()` must be executed from the command line. The above program can be executed within Jupyter:

```python
!python3 -m riscemu printsum.s
```
